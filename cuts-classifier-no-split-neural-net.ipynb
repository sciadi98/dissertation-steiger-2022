```python
# Imports and global options
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout

pd.set_option('display.max_columns', None)
```

```python
import os
os.getcwd()
```

```python
# Load data
df = pd.read_excel('DATA/Train/Final_dataset.xlsx')
df.head()
```

```python
df.columns
```

```python
# Filter bad data, replace categories with dummies
data = pd.get_dummies(df, columns =['shape', 'profile_entrance', 'profile_exit', 'rising_entrance', 'rising_exit', 'shards', 'fethering', 'entrance_mounding', 
                                    'center_mounding', 'exit_mounding', 'mounding', 'result'], drop_first=True)
data.head()
```

```python
# Extract data for model and do train-test split
x_train = data.iloc[ : , 8:-1]
y_train = data.iloc[ : , -1]
```

```python
# Instantiate the model
model = Sequential()
model.add(Dense(16, activation='relu', input_shape=(x_train.shape[1],)))
model.add(Dropout(0.2))
model.add(Dense(16, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train (=fit) the model
model.fit(x_train, y_train, epochs=100, batch_size=20, validation_split=0.2)
```

```python
model.get_params()
```

```python
# Load test data
df_test = pd.read_excel('DATA/Validation/Dataset_da_validare.xlsx')
df_test.head()
```

```python
# Filter bad data, replace categories with dummies
data_test = pd.get_dummies(df, columns =['shape', 'profile_entrance', 'profile_exit', 'rising_entrance', 'rising_exit', 'shards', 'fethering', 'entrance_mounding', 
                                    'center_mounding', 'exit_mounding', 'mounding', 'result'], drop_first=True)
data_test.head()
```

```python
# Extract data for model
x_test = data_test.iloc[ : , 8:-1]
y_test = data_test.iloc[ : , -1]
```

```python
# Validate the model
model.predict(x_test)
```
