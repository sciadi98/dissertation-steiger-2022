```python
# Imports and global options
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

pd.set_option('display.max_columns', None)
```

```python
import os
os.getcwd()
```

```python
# Load data
df = pd.read_excel('DATA/Train/Final_dataset.xlsx')
df.head()
```

```python
df.columns
```

```python
# Filter bad data, replace categories with dummies
data = pd.get_dummies(df, columns =['shape', 'profile_entrance', 'profile_exit', 'rising_entrance', 'rising_exit', 'shards', 'fethering', 'entrance_mounding', 
                                    'center_mounding', 'exit_mounding', 'mounding', 'result'], drop_first=True)
data.head()
```

```python
# Extract data for model and do train-test split
x_train = data.iloc[ : , 8:-1]
y_train = data.iloc[ : , -1]
```

```python
# Instantiate the model
model = LogisticRegression(solver='liblinear', random_state=0)
```

```python
# Train (=fit) the model
model.fit(x_train, y_train)
```

```python
model.get_params()
```

```python
# Load test data
df_test = pd.read_excel('DATA/Validation/Dataset_da_validare.xlsx')
df_test.head()
```

```python
# Filter bad data, replace categories with dummies
data_test = pd.get_dummies(df, columns =['shape', 'profile_entrance', 'profile_exit', 'rising_entrance', 'rising_exit', 'shards', 'fethering', 'entrance_mounding', 
                                    'center_mounding', 'exit_mounding', 'mounding', 'result'], drop_first=True)
data_test.head()
```

```python
# Extract data for model and do train-test split
x_test = data_test.iloc[ : , 8:-1]
y_test = data_test.iloc[ : , -1]
```

```python
# Validate the model
model.predict(x_test)
```

```python
model.score(x_test, y_test)
```
