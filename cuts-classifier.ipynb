```python
# Imports and global options
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

pd.set_option('display.max_columns', None)
```

```python
import os
os.getcwd()
```

```python
# Load data
df = pd.read_excel('Final_dataset.xlsx')
df.head()
```

```python
df.columns
```

```python
# Filter bad data, replace categories with dummies
data = pd.get_dummies(df, columns =['shape ', 'profile entrance ', 'profile exit ', 'rising entrance vs1', 'rising exit vs2', 'shards', 'fethering', 'entrance mounding vs', 
                                    'center mounding vs.2 ', 'exit mounding vs', 'mounding', 'result'], drop_first=True)
data.head()
```

```python
# Extract data for model and do train-test split
x = data.iloc[ : , 8:-1]
y = data.iloc[ : , -1]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=0)
#x.drop(x.columns[8], axis=1, inplace=True)
x
```

```python
# Instantiate the model
model = LogisticRegression(solver='liblinear', random_state=0)
```

```python
# Train (=fit) the model
model.fit(x_train, y_train)
```

```python
# Validate the model
model.predict(x_test)
```

```python
# Show expected results
y_test
```

```python
model.score(x_test, y_test)
```
